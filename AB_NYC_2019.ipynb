# IDENTIFY OUTLIERS

### Identify Outlier in latitude Column

```python
import matplotlib.pyplot as mtl
import seaborn as sns

mtl.figure(figsize = (3,4))
sns.boxplot(x = "latitude", data = dataset)
mtl.show()
```

### Identify Outlier in longitude Column

```python
import matplotlib.pyplot as mtl
import seaborn as sns

mtl.figure(figsize = (3,4))

mtl.boxplot("longitude" , data = dataset , showmeans = True)
mtl.show()
```
